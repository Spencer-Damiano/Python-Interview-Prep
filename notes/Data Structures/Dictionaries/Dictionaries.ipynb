# Dictionaries

- How to define dictionaries and their methods. For their use in algorithms see the solutions folder
- Dictionaries are ordered, allowing them to be used by their key
- Dictionaries are changeable which allows us to add or remove items after it has been created.

## Defining a dictionary

```python
sample_dictionary = {
    "dicts": int(1),
    "can": str(2),
    "have": bool(0),
    "any": list(range(1,5)),
    "types": {"another": "dict"}
}
```

## Dictionary Keys

```python

print(sample_dictionary.keys())# this gets you a list of the keys `dict_keys(['dicts', 'can', 'have', 'any', 'types'])`

sample_dictionary["new"] = None # there has to be a placeholder value when creating a new key
```

## Dictionary Values

```python

print(sample_dictionary.values())# this gets you a list of the keys `dict_keys(['dicts', 'can', 'have', 'any', 'types'])`

sample_dictionary["new"] = "Changed" # Any value in a dictionary can be changed

print(sample_dictionary.values())
```
